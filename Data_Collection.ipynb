<h3><b>Install & Import</b></h3>

Cài đặt và import những thư viện cần thiết cho phần thu thập dữ liệu.

```python
#Import các thư viện
import requests
import csv
```

```python
total = 9000
offset = 100
pages = total // offset

job_data = []

endpoint = "https://ms.vietnamworks.com/job-search/v1.0/search"
payload = {
    "hitsPerPage": offset,
    # "page": 1,
    "retrieveFields": [
        "benefits",
        "jobTitle",
        "salaryMax",
        "isSalaryVisible",
        "jobLevelVI",
        "isShowLogo",
        "salaryMin",
        "companyLogo",
        "userId",
        "jobLevel",
        "jobId",
        "companyId",
        "approvedOn",
        "isAnonymous",
        "alias",
        "expiredOn",
        "industries",
        "workingLocations",
        "services",
        "companyName",
        "salary",
        "onlineOn",
        "simpleServices",
        "visibilityDisplay",
        "isShowLogoInSearch",
        "priorityOrder",
        "skills",
    ],
}
```

```python
for x in range(pages):
    payload["page"] = x + 1
    print("Page: ", x + 1)
    resp = requests.post(endpoint, json=payload)
    # print(x.json()["data"])
    data = resp.json()
    job_data = job_data + data["data"]
```

```python
# now we will open a file for writing
data_file = open("data/job_list.csv", "w",encoding='UTF8', newline='')

# create the csv writer object
csv_writer = csv.writer(data_file)

# Counter variable used for writing
# headers to the CSV file
count = 0

for job in job_data:
    if count == 0:
        # Writing headers of CSV file
        header = job.keys()
        csv_writer.writerow(header)
        count += 1

        # Writing data of CSV file
    csv_writer.writerow(job.values())


data_file.close()
```
